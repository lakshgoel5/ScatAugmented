```python
def parse_bcch_bch_message(binary_str):
    if len(binary_str) < 24:
        raise ValueError("Expected at least 24 bits for BCCH-BCH message")

    choice_bit = binary_str[0]

    if choice_bit == '0':
        # It's a MIB
        mib_bits = binary_str[1:24] 
        return {
            "type": "mib",
            "decoded": parse_mib(mib_bits)
        }

    elif choice_bit == '1':
        return {
            "type": "messageClassExtension",
            "decoded": "Not implemented (usually empty)"
        }

    else:
        raise ValueError("Invalid CHOICE bit")

def parse_mib(bits):
    def bits_to_int(b): return int(b, 2)

    return {
        "systemFrameNumber": bits_to_int(bits[0:6]),
        "subCarrierSpacingCommon": {
            "0": "scs15or60",
            "1": "scs30or120"
        }[bits[6]],
        "ssb-SubcarrierOffset": bits_to_int(bits[7:11]),
        "dmrs-TypeA-Position": {
            "0": "pos2",
            "1": "pos3"
        }[bits[11]],
        "controlResourceSetZero": bits_to_int(bits[12:16]),
        "searchSpaceZero": bits_to_int(bits[16:20]),
        "cellBarred": {
            "0": "barred",
            "1": "notBarred"
        }[bits[20]],
        "intraFreqReselection": {
            "0": "allowed",
            "1": "notAllowed"
        }[bits[21]],
        "spare": bits[22:23],
    }
```
